```python
# # 創database

# import csv
# import MySQLdb
# import pandas as pd
# import xlrd
# from datetime import datetime

# conn = MySQLdb.connect(host="127.0.0.1",    # your host, usually localhost
#                      user="root",         # your username
#                      passwd="",           # your password
# #                      db = "top_line_exam",    # name of the data base         
#                      use_unicode=True, 
#                      charset="utf8")
# cur = conn.cursor()
# cur.execute("CREATE DATABASE top_line_exam DEFAULT CHARACTER SET 'utf8'")
# conn.commit()    
# cur.close()
# conn.close()




# 在某一個DB創一個新的table

import csv

import MySQLdb
import pandas as pd
import xlrd
from datetime import datetime
# cur.execute("CREATE DATABASE top_line_exam DEFAULT CHARACTER SET 'utf8'")

conn = MySQLdb.connect(host="127.0.0.1",    # your host, usually localhost
                     user="root",         # your username
                     passwd="",           # your password
                     db = "top_line_exam",    # name of the data base         
                     use_unicode=True, 
                     charset="utf8")


cur = conn.cursor()

# ---------------------------------------------------------------------------------------------

cur.execute("DROP TABLE IF EXISTS Items")
# create table & columns & datatype of each column --------------- 關於日期部份，不能有"-" : 2017-01-01 -> 20170101
sql = """CREATE TABLE Items (
         Item_ID VARCHAR(5),
         Name VARCHAR(5),
         Cuft FLOAT(5),
         Manufacturer VARCHAR(5)
         );"""         
cur.execute(sql)
# insert into SQL
sql = """INSERT INTO Items (Item_ID, Name, Cuft, Manufacturer) 
         VALUES ("001", "A", 1.2, "W"), 
                ("002", "B", 2.3, "X"), 
                ("003", "C", 3.4, "Y"), 
                ("004", "D", 4.5, "Z");"""
cur.execute(sql)

# -------------

cur.execute("DROP TABLE IF EXISTS Item_Location_Map")
# create table & columns & datatype of each column --------------- 關於日期部份，不能有"-" : 2017-01-01 -> 20170101
sql = """CREATE TABLE Item_Location_Map (
         Item_ID VARCHAR(5),
         Location_ID VARCHAR(5),
         Available_Count INT(5)
         );"""         
cur.execute(sql)
# insert into SQL
sql = """INSERT INTO Item_Location_Map (Item_ID, Location_ID, Available_Count) 
         VALUES ("001", "M", 15), 
                ("001", "N", 3), 
                ("002", "M", 20), 
                ("002", "N", 36),
                ("003", "M", 2),
                ("004", "N", 1);"""
cur.execute(sql)

# -------------

cur.execute("DROP TABLE IF EXISTS Locations")
# create table & columns & datatype of each column --------------- 關於日期部份，不能有"-" : 2017-01-01 -> 20170101
sql = """CREATE TABLE Locations (
         Location_ID VARCHAR(5),
         Name VARCHAR(5)
         );"""         
cur.execute(sql)
# insert into SQL
sql = """INSERT INTO Locations (Location_ID, Name) 
         VALUES ("M", "US"), 
                ("N", "TW");"""
cur.execute(sql)

# ---------------------------------------------------------------------------------------------
conn.commit()    


cur.close()
conn.close()
```

```python
# 2019.01.08 ver_2 (修正)


# 啟用driver，連結selenium和chrome瀏覽器
driver_path = "/home/js/py_ds_nas/chromedriver" # 改為你的 driver 路徑
driver = webdriver.Chrome(driver_path)


# 用chrome打開網站
driver.get("https://www.overstock.com/Home-Garden/Living-Room-Furniture/713/cat.html")

# 點對話框
search_elem = driver.find_element_by_css_selector("#search-input")
time.sleep(3)

search_elem = driver.find_element_by_css_selector(".cookie-consent-accept")
time.sleep(3)


# 想要往下捲到底 : 按END鍵 
# Change the option
actions = ActionChains(driver)

count = 0

while count < 3 :  #5

    actions.send_keys(Keys.END)
    actions.perform()
    time.sleep(2)

    count = count + 1
    print(count)

    
    

#設置假的瀏覽器資訊
import requests
headers = requests.utils.default_headers()
headers['User-Agent'] = 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36'
# r = requests.get('https://www.overstock.com/Home-Garden/Living-Room-Furniture/713/cat.html', headers=headers)
  
    
    
url = ("https://www.overstock.com/Home-Garden/Living-Room-Furniture/713/cat.html")

resp = requests.get(url, headers=headers)

PageSource = driver.page_source
soup = BeautifulSoup(PageSource, 'lxml')    # resp.text / PageSource


print("\n")

name_list = []
price_list = []
reviews_list = []
stars_list = []




#------

# content = soup.find_all('div', id=re.compile('product-container'))
for i in range(0, 60):


    content = soup.find_all('div', id=re.compile('product-container'))[0] # 不然前面有4個出清商品在spotfront-wrapper

    main_info = content.find('div', id=re.compile('product-%s' % i))
#     name
    # name = name.find('div', class_ = 'product-info').find('div', class_ = 'product-title medium-title').text
    try :
        if main_info.find('div', class_ = 'product-title').text == None :
            name = main_info.find('div', class_ = 'product-title medium-title').text
        else :
            name = main_info.find('div', class_ = 'product-title').text
    except :
        name = main_info.find('div', class_ = 'product-title medium-title').text

    name_list.append(name)
    print(name)
#     price
    try :        
        price = content.find_all('div', class_ = 'product-info')[i].find('div', class_ = 'product-price sale').text
    except  :
        try :
            price = main_info.find('div', class_ = 'product-price standard').text
        except :
            price = main_info.find('div', class_ = 'product-price see-price-in-cart').text 
#     except :
#         None
    price_list.append(price)
    print(price)
#     reviews
    try :
        reviews = main_info.find('div', class_ = 'product-footer').find('span', class_ = 'product-review').text
    except :
        reviews = None
    reviews_list.append(reviews)
    print(reviews)
#     stars
    try :
        stars = main_info.find('div', class_ = 'product-footer').find('span', class_ = 'product-review').parent.find('div', class_ = 'reviews').attrs['title']
    except :
        stars = None
    stars_list.append(stars)
    print(stars)

# 資訊重新歸零
    main_info = None




driver.close()

# print(name_list)
# print(price_list)
# print(reviews_list)
# print(stars_list)

df = pd.DataFrame({"name" : name_list, "price" : price_list, "stars" : stars_list, "reviews" : reviews_list})
df
```
