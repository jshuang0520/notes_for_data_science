```python
# Google map demo code - 2018.12.13 更新版本 : aria-label位置有變，268行那邊有改span tag的位置


# 2018.03.21 - 成功!

# 版本 30 - 把<可能造成 ElementNotVisibleException 的原因排除 >
# 1. css_selector會選到兩個以上 : 改用xpath
# 2. 把視窗移到可以看見 要讓selenium按的按鈕那裡


# http://www.compjour.org/warmups/govt-text-releases/intro-to-bs4-lxml-parsing-wh-press-briefings/
# https://www.slideshare.net/tw_dsconf/python-78691041

# https://stackoverflow.com/questions/8049520/web-scraping-javascript-page-with-python

# Is there a way to use PhantomJS in Python?
# https://stackoverflow.com/questions/13287490/is-there-a-way-to-use-phantomjs-in-python

def google_map_info(x):

    '''
    請在您輸入關鍵字的前後加上引號 ( " " )
    
    範例 : 
    google_map_info("宋廚")
    
    
    初次使用時，請更新您的chrome瀏覽器 :
    https://www.google.com/chrome/browser/desktop/index.html

    並請至以下網址安裝chromedriver :
    https://sites.google.com/a/chromium.org/chromedriver/downloads
    
    最後將程式內文 driver_path 改為您的 driver 路徑
    ex. driver_path = "C:/Users/JS/Downloads/chromedriver_win32/chromedriver.exe" # 改為你的 driver 路徑
    '''
    import numpy as np
    import pandas as pd
    
    from selenium import webdriver
    from selenium.webdriver import ActionChains
    from selenium.webdriver.common.keys import Keys
    import time
    # 如果有 NoSuchElementException 的情況
    from selenium.common.exceptions import NoSuchElementException
      
    from bs4 import BeautifulSoup
    import requests

    import datetime
    now = datetime.datetime.now()


    # 啟用driver，連結selenium和chrome瀏覽器
    driver_path = "/home/js/py_ds_nas/chromedriver" # 改為你的 driver 路徑
#     driver_path = "C:/Users/JS/Downloads/chromedriver_win32/chromedriver.exe" # 改為你的 driver 路徑
#     driver_path = "C:/Users/johnson_huang.EOL/Downloads/chromedriver" # 改為你的 driver 路徑
#     driver_path = "C:/Users/bigdata01/Downloads/chromedriver" # 改為你的 driver 路徑
    driver = webdriver.Chrome(driver_path)
    


#     # 啟用driver，連結selenium和 Firefox 瀏覽器  
#     driver_path = "C:/Users/bigdata01/Downloads/geckodriver-v0.20.0-win64/geckodriver" # 改為你的 driver 路徑
#     driver = webdriver.Firefox(executable_path = driver_path)
    
    
    # 用chrome打開google map的網站
    driver.get("https://www.google.com.tw/maps/@23.4857501,120.0843006,7z?hl=zh-TW")
    
    # 點對話框
    search_elem = driver.find_element_by_css_selector("#searchboxinput")
    time.sleep(3)
    
    # 先把input轉成文字，再輸入
    x = str(x)
    search_elem.send_keys(x)
    time.sleep(5)
    
    # 按放大鏡搜尋
    search_button_elem = driver.find_element_by_css_selector("#searchbox-searchbutton")
    search_button_elem.click()
    time.sleep(5)
    
    
    
    # 進網頁部份 - 分三種情況 1.沒連鎖店 2.有連鎖店(會先選分店再進來) 3.學校(沒有營業時間的收合按鈕)
    
    try :
        # 把營業時間展開
        time.sleep(2)
        search_time_button_elem_1 = driver.find_element_by_css_selector(".maps-sprite-pane-info-arrowup")    
        time.sleep(1)
        search_time_button_elem_1.click()
        time.sleep(2)  
    
    # 其他狀況就會有 NoSuchElementException
    except NoSuchElementException :
        time.sleep(2)
        # 2. 如果是連鎖店，要再多一層 (pizza hut)
        try :
            time.sleep(2)
            # 先點選分店 - 選擇第1間 child(1)
            search_branches_elem = driver.find_element_by_css_selector(".section-result:nth-child(1) .section-result-title span")
            search_branches_elem.click()
            time.sleep(2) 
            
            # 把營業時間展開
            time.sleep(2)
            search_time_button_elem_1 = driver.find_element_by_css_selector(".maps-sprite-pane-info-arrowup")    
            time.sleep(1)
            search_time_button_elem_1.click()
            time.sleep(2)  
        # 3. 如果都不是以上狀況(沒有分店可以選，也沒有營業時間可以展開)    
        except NoSuchElementException :
                    time.sleep(3)

    # 知道他有幾篇評論
#     number_of_comments = driver.find_element_by_css_selector(".section-reviewchart-numreviews")
    number_of_comments = driver.find_element_by_xpath('//*[@id="pane"]/div/div[1]/div/div/div[1]/div[3]/div[2]/div/div[1]/span[3]/ul/li/span[2]/span[1]/button')
    time.sleep(1) 
    number_of_comments = number_of_comments.text
    time.sleep(1)
    number_of_comments = number_of_comments.replace("," , "")
    time.sleep(1)
    number_of_comments = number_of_comments.replace("篇評論" , "")
    time.sleep(1)
    number_of_comments = number_of_comments.replace(" " , "")
    time.sleep(1)
    number_of_comments = number_of_comments.replace("'" , "")
    time.sleep(1)
    number_of_comments = number_of_comments.replace('"' , "")
    time.sleep(1)
    number_of_comments = int(number_of_comments)
    time.sleep(2)
                
    
    # 想要往下捲到底 : 按END鍵 
    # Change the option
    actions = ActionChains(driver)

    count = 0

    while count < 2 :

        actions.send_keys(Keys.END)
        actions.perform()
        time.sleep(2)

        count = count + 1
    
    
    # 想要往上捲到底 : 按 ARROW_UP 鍵 
    # Change the option
    actions = ActionChains(driver)

    count = 0

    while count < 3 :
        actions.send_keys(Keys.PAGE_UP)
        actions.perform()
        
        count = count + 1
    
    time.sleep(2)
    
    # parse - text
    
    # 抓第一層資訊 : 
    first_layer_elem = driver.find_elements_by_css_selector(".section-divider+ .section-info-hoverable .section-info-line , .section-reviewchart-numreviews , .section-info-hoverable:nth-child(4) .widget-pane-link , .section-info-speak-numeral .widget-pane-link , .section-subheader:nth-child(22) .section-subheader-large-text , .section-review-snippet-line , .section-rating-term .widget-pane-link , .widget-pane-info-open-hours-row-interval li , .widget-pane-info-open-hours-row-header div , .section-info-red , .section-star-display , .section-hero-header-title")
    first_layer_elem_output = [first_result.text for first_result in first_layer_elem]

    time.sleep(2)
  



    # 抓第二層資訊 :
    
    # ---------- 第二層階段1開始 : 進入 所有使用者評論 的頁面 ---------- 
    time.sleep(4)
#     comments_of_宋廚 = driver.find_element_by_css_selector(".section-reviewchart-numreviews")
#     comments_of_宋廚 = driver.find_element_by_xpath("//*[@id='pane']/div/div[1]/div/div/div[22]/div/button")
    comments_of_宋廚 = driver.find_element_by_xpath('//*[@id="pane"]/div/div[1]/div/div/div[1]/div[3]/div[2]/div/div[1]/span[3]/ul/li/span[2]/span[1]/button')
    time.sleep(2)
    comments_of_宋廚.click()
    time.sleep(2) 

    # 準備抓取評論 :

    # 先點擊左邊螢幕空白處(把滑鼠移到左邊)
    left = driver.find_element_by_css_selector(".section-star-display")
    time.sleep(1)
    left.click()
    time.sleep(1)    

    # 點擊左方空白處
    left = driver.find_element_by_css_selector(".section-review-content")
    time.sleep(1)
    left.click()
    time.sleep(2)

    # 想要把評論往下捲到底 : 按END鍵 
    # Change the option
    actions = ActionChains(driver)

    count = 0
    
    # 先定義要按幾次END鍵
    if number_of_comments == 0 :
        n = 1
    elif (0 < number_of_comments and number_of_comments < 100 ) :
        n = 20
    elif (100 < number_of_comments and number_of_comments < 400 ) :
        n = 40
    elif (400 < number_of_comments and number_of_comments < 700 ) :
        n = 50
    elif (700 < number_of_comments and number_of_comments < 840 ) :
        n = 60
    elif (840 < number_of_comments and number_of_comments < 980 ) :
        n = 70
    elif (980 < number_of_comments and number_of_comments < 1120 ) :
        n = 80
    elif (1120 < number_of_comments and number_of_comments < 1260 ) :
        n = 90
    elif (1260 < number_of_comments and number_of_comments < 1400 ) :
        n = 100
    else :
        n = (number_of_comments / 15)
        

    while count < n :

        actions.send_keys(Keys.END)
        actions.perform()
        time.sleep(2)

        count = count + 1



    # 開始抓取評論  //div/div[@class='section-review-content']
    third_result_comments = driver.find_elements_by_xpath("//div/div[@class='section-review-content']")
    # .text()解析文字
    third_result_comments_output = pd.Series([third_result.text.replace("★★★★★" , "") for third_result in third_result_comments])
    
    time.sleep(10)

    
    # ----- 抓取星等數 & 解析 -----
    
    # get current url
    comment_url = driver.current_url
    
    # request
#     response = requests.get(comment_url, cookies = {'over18': '1'}) # 用 requests 的 get 方法把網頁抓下來
    PageSource = driver.page_source
    
    # parse
#     html_doc = response.text   # text 屬性就是 html 檔案
#     soup = BeautifulSoup(response.text, "lxml")  # type(soup) == bs4.BeautifulSoup
    soup = BeautifulSoup(PageSource, "lxml") # 用 driver.page_source 給 BeautifulSoup解析 # type(soup) == bs4.BeautifulSoup
    # 等待網頁解析
    time.sleep(20)
    
    # 抓取星等
    ratings = [] 
    for h in soup.find_all('div', class_ = "section-review-metadata section-review-metadata-with-note"):
#         a = h.find('span')     ## 因為好像改版了，不在第一個span裡面
        a = h.find_all('span')[1]
        ratings.append(a.attrs['aria-label'])
    # 等他抓資料    
    time.sleep(60)
    

    # ---------- 第二層階段1結束  ---------- 

#     out = [first_layer_elem_1, first_layer_elem_2, first_layer_elem_3, first_layer_elem_4, first_layer_elem_5, first_layer_elem_6, third_result_comments_output]
#     out.to_csv("%d.%d.%d %d點%d分 google map - %s.csv" % (now.year, now.month, now.day, now.hour, now.minute, x))


    df_第一層 = pd.DataFrame(first_layer_elem_output, columns=["第一層_基本資料"])
    
    df_第二層 = pd.DataFrame(third_result_comments_output, columns=["第二層_所有評論"])
    df_第二層['使用者評價星等'] = ratings

    # 去除重複的row
    df_not_empty_row_filter = df_第一層["第一層_基本資料"] != ""
    df_第一層 = df_第一層[df_not_empty_row_filter]
    df_第一層 = df_第一層.reset_index(drop=True)
    
    # 寫成 excel檔，輸出
    writer = pd.ExcelWriter('%d.%d.%d %d點%d分 google map - %s.xlsx' % (now.year, now.month, now.day, now.hour, now.minute, x), engine='xlsxwriter')
    df_第一層.to_excel(writer, sheet_name='第一層')
    df_第二層.to_excel(writer, sheet_name='第二層')


    driver.close()
    
    return df_第一層, df_第二層
```
